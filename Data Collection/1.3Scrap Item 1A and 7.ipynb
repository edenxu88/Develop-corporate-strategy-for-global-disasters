```python
def Item1A(path):
    htmlfile = open(path, 'r', encoding='utf-8')
    htmlhandle = htmlfile.read()
    htmlfile.close()
    soup = bs(htmlhandle,'html.parser')
    re1 = r'(?<=(?i)ITEM\xa01A. RISK FACTORS).*?(?=(?i)ITEM\xa01B)'
    re2 = r'(?<=(?i)ITEM 1A.RISK FACTORS).*?(?=(?i)ITEM 1B.Unresolved Staff Comments)'
    re3 = r'(?<=(?i)ITEM 1A. RISK FACTORS).*?(?=(?i)ITEM 1B. UNRESOLVED STAFF COMMENTS)'
    
    IA1 = []
    re_all=[re1,re2]
    for i in re_all:
        Item1Arow = re.findall(i,soup.text)
        IA1 = IA1+Item1Arow
    filter(None, IA1)
    
#     for i in IA1:
#         IA1.replace('\xa0',' ')
    
    return IA1
```
